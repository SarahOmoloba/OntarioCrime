### Crime Statistics

```python
# Import the data from 'statistics.csv'
crime_statistics_df = pd.read_csv('ReducedCrime2016Cleaned.csv')
crime_statistics_df.head(5)
```

```python
crime_statistics_df.shape
```

```python
crime_stats_df = crime_statistics_df.drop(['DGUID','UnitOfMeasurement','UOM_ID', 'SCALAR_FACTOR','VECTOR','COORDINATE','DECIMALS'], axis = 1)
```

```python
crime_stats_df.dtypes
```

## Merge the tables together

#### Crime Statistics and Statistics

```python
stats_join_crime_stats_df = pd.merge(crime_stats_df, statistics_df, on='statistic', how ='left')
```

```python
stats_join_crime_stats_df.head()
```

```python
crime_stats_df['statId'] = stats_join_crime_stats_df['statId']
crime_stats_df.head()
```

#### Crime Statistics and Geo2

```python
geo_join_crime_stats_df = pd.merge(crime_stats_df, geo2_df, on='geo', how ='left')
geo_join_crime_stats_df.head()
```

```python
crime_stats_df['geoID'] = geo_join_crime_stats_df['geoID']
crime_stats_df.head()
```

```python
# change the data type of the geoID to int
crime_stats_df['geoID'] = crime_stats_df['geoID'].astype(int)
crime_stats_df.head()
```

```python
# Save the data as 'crimestatmerged.csv'
#crime_stats_df.to_csv('crime_stat_merged.csv', index = False)
```

#### Crime Statistics and Violations

```python
violation_join_crime_stats_df = pd.merge(crime_stats_df, violations_df, on='violation', how ='left')
violation_join_crime_stats_df.tail()
```

```python
violation_join_crime_stats_df.isnull().sum()
```

```python
crime_stats_df['violationId'] = violation_join_crime_stats_df['violationId']
crime_stats_df.head()
```

```python
crime_stats_df.tail()
```

```python
crime_stats_df['violationId'].fillna(0, inplace=True)
crime_stats_df.head()
```

```python
crime_stats_df['violationId'] = crime_stats_df['violationId'].astype(int)
crime_stats_df.head()
```
